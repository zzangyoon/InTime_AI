```python
import pandas as pd

df = pd.DataFrame(rows, columns=[
    "employee_id", "name", "check_date", "check_in", "check_out", "check_in_method", "check_out_method", "memo"
])

df
```

### Df merge

```python
df_checkin = (
    df[df["check_in"].notna()]
    .sort_values(["employee_id", "check_date", "check_in"])
    .groupby(["employee_id", "check_date"], as_index = False)
    .first()
)
df_checkin
```

```python
df_checkout = (
    df[df["check_out"].notna()]
    .sort_values(["employee_id", "check_date", "check_out"])
    .groupby(["employee_id", "check_date"], as_index = False)
    .last()
)
df_checkout
```

```python
df_merged = pd.merge(
    df_checkin,
    df_checkout,
    on = ["employee_id", "check_date"],
    suffixes = ("_in", "_out"),
    how = "outer"
)
df_merged
```

```python
# merge 후 컬럼 정리
df_final = df_merged[[
    "employee_id",
    "name_in",
    "check_date",
    "check_in_in",
    "check_out_out",
    "check_in_method_in",
    "check_out_method_out",
    "memo_in",
    "memo_out"
]]
df_final
```

```python
df_final = df_final.rename(columns={
    "name_in" : "name",
    "check_in_in" : "check_in",
    "check_out_out" : "check_out",
    "check_in_method_in" : "check_in_method",
    "check_out_method_out" : "check_out_method"
})
df_final
```

```python
# automation/services/attendance_reporter.py

from automation.db.db_access import get_attendance_data
from automation.llm.llm_client import summarize_attendance  # (LLM 인터페이스)
import pandas as pd
from datetime import datetime

def generate_attendance_report(date: str = None) -> str:
    """
    주어진 날짜의 근태 데이터를 요약 보고서 형태로 생성
    """
    # 1. 날짜 기본값: 오늘
    if date is None:
        date = datetime.now().strftime("%Y-%m-%d")

    # 2. DB에서 데이터 조회
    rows = get_attendance_data(date)

    if not rows:
        return f"{date}의 근태 데이터가 없습니다."

    # 3. DataFrame으로 변환
    df = pd.DataFrame(rows, columns=[
        "employee_id", "name", "check_date", "check_in", "check_out",
        "check_in_method", "check_out_method", "memo"
    ])

    # 4. 필요한 전처리 / 가공 (예: 지각자 추출 등)
    # 예: 출근 시간 비교해서 지각 여부 판단
    df["is_late"] = df["check_in"].apply(lambda t: t > f"{date} 09:00:00")

    # 5. 보고서용 텍스트 생성 (LLM 입력용)
    summary_prompt = create_summary_prompt(df)

    # 6. LLM에 요약 요청
    summary = summarize_attendance(summary_prompt)

    return summary


def create_summary_prompt(df: pd.DataFrame) -> str:
    """
    LLM에 전달할 수 있도록 근태 데이터를 텍스트로 포맷팅
    """
    lines = []

    for _, row in df.iterrows():
        line = f"{row['name']} (사번 {row['employee_id']}): 출근 {row['check_in']}, 퇴근 {row['check_out']}"
        if row["is_late"]:
            line += " → 지각"
        if row["memo"]:
            line += f" (비고: {row['memo']})"
        lines.append(line)

    report = "\n".join(lines)
    return f"""
다음은 {df.iloc[0]['check_date']}의 사원 근태 기록입니다.

{report}

이 데이터를 바탕으로 요약된 보고서를 작성해 주세요.
"""

```

```python
import ollama

model = "gemma3:4b"

system_prompt = """
당신은 회사의 인사 담당자입니다.
다음 근태 데이터를 바탕으로 하루동안의 직원 출근/퇴근 및 특이사항을 보고서로 작성하는 역할을 합니다.
결근자나 지각자는 강조해주세요.
"""

user_prompt = """
다음은 근태 데이터입니다.

employee_id : 사원번호,
check_date : 근무 날짜,
name : 이름,
check_in : 출근 시간,
check_out : 퇴근 시간,
check_in_method : 출근 방법(FD, FR : 얼굴인식 / ID : 사원증),
check_out_method : 퇴근 방법(FD, FR : 얼굴인식 / ID : 사원증),
memo_in	: 비고,
memo_out : 비고,
is_late : 지각 유무 (T/F)

근태 보고서를 만들어주세요.
"""

response = ollama.chat(
    model = model,
    messages = [
        {"role" : "system", "content" : system_prompt},
        {"role" : "user", "content" : user_prompt}
    ]
)

print(response)
```
